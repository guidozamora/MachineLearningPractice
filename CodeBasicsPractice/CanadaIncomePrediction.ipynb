# Linear Regression Model Excercise
Linear Regression Model to predict Canada per capita Income in year 2020

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
income = pd.read_csv('canada_per_capita_income.csv')
income.columns = ['year', 'income']
income
```

```python
%matplotlib inline
plt.xlabel('Year', fontsize=15)
plt.ylabel('Income Per capita',fontsize=15)
plt.scatter(income.year,income.income,color='red',marker='*')
```

```python
reg = linear_model.LinearRegression()
reg.fit(income[['year']],income.income)
```

```python
reg.predict([[2020]])
```
